# Reading data into SQLite database

```python
import pandas as pd                 # Import Pandas library
import sqlite3                        # Import SQLite3 library 
con = sqlite3.connect('northwind.db') # Connect to SQLite database
                                      # Assign SQL variable
sql = "SELECT customerID, companyName, address, city, postalCode FROM customers WHERE city = 'London'" 
df = pd.read_sql_query(sql,con)                         # Read result into dataframe
con.close()                                             # Close connection
df                                                      # Display data frame
```

```python
import pandas as pd
import sqlite3
con = sqlite3.connect('northwind.db')
sql = "SELECT * FROM products WHERE unitPrice BETWEEN 19 AND 20"
df = pd.read_sql_query(sql,con)
con.close()
df
```

```python
import pandas as pd
import sqlite3
con = sqlite3.connect('northwind.db')
sql = "SELECT categoryID, ROUND(sum(unitsInStock),2) AS [Total Units in Stock] FROM products  GROUP BY categoryID"
df = pd.read_sql_query(sql,con)
con.close()
df
```

```python
import pandas as pd
import sqlite3
con = sqlite3.connect('northwind.db')
s1 = "SELECT Orders.OrderID, Customers.contactName, Customers.contactTitle, Customers.companyName, Orders.OrderDate "
s2 = "FROM Orders INNER JOIN Customers ON Orders.CustomerID=Customers.CustomerID "
sql = s1 + s2
df = pd.read_sql_query(sql,con)
con.close()
df
```

```python
import pandas as pd
import sqlite3
con = sqlite3.connect('northwind.db')
s1 = "SELECT Orders.OrderID, order_details.productID, products.productName, Order_details.quantity, Order_details.unitPrice "
s2 = "FROM Orders "
s3 = "INNER JOIN  Order_details ON Orders.orderID = Order_details.orderID "
s4 = "INNER JOIN Products ON Order_details.productID = Products.productID "
sql = s1 + s2 + s3 + s4
df = pd.read_sql_query(sql,con)
con.close()
df
```

```python
import pandas as pd
import sqlite3
con = sqlite3.connect('nba_db.db')
sql = "SELECT Col_College, sum(Col_Game) AS [Total # of Games] FROM nba_run_results GROUP BY Col_College ORDER BY [Total # of Games] DESC "
df = pd.read_sql_query(sql,con)
con.close()
df
```
